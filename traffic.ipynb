```python
%%writefile app.py
import streamlit as st
import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def get_data(query, params=None):
    conn = mysql.connector.connect(user='root', password='agila1991', host='localhost', database='project1')
    if params:
        df = pd.read_sql_query(query, conn, params=params)
    else:
        df = pd.read_sql_query(query, conn)
    conn.close()
    return df

# Streamlit App Title
st.set_page_config(page_title="POLICE CHECK POST DATA ANALYSIS", layout="wide") 

# Sidebar for navigation
st.sidebar.title("Navigation")
page = st.sidebar.radio("Go to", ["Project Introduction", "Check post Data Visualization", "Prediction","SQL Queries", "Creator Info"])

# -------------------------------- PAGE 1: Introduction --------------------------------

if page == "Project Introduction":
    
    st.title("🚦 Check Post Data Analysis")
    st.image("A:/Checkpost1.jpg")
    st.subheader("A Streamlit App for Exploring Violation Trends")

    st.markdown("""
        <div style='color: #2E8B57;'> 
        This project focuses on the analysis of Police Check Post data from different countries, utilizing an SQL database 
        to extract meaningful insights and trends. By examining data related to traffic violations, drug-related stops,
        and other relevant parameters.
        </div>
    """, unsafe_allow_html=True)

    st.markdown("""
        <div style='color: #4682B4;'>  
        <strong>Features:</strong>
        <ol>
            <li><strong>Data Filtering:</strong> 
            Users can view and filter Check Post data by various criteria, including country, stop date, stop time, driver age, 
            and race. This feature allows for a more granular analysis of the data and helps identify specific trends 
            within different demographics.  
            </li>
            <li><strong>Dynamic Visualizations 🔮:</strong> 
             🚀The project provides interactive visualizations that represent the data in a user-friendly manner. These visualizations
            help to view  the trends related to traffic violations and enforcement actions.
            </li>
            <li><strong>Predefined SQL Queries:</strong>
            Users can run a set of predefined SQL queries to explore the data further.📚These queries are designed to extract specific insights
            such as the frequency of drug-related stops or the correlation between driver demographics and violation rates.
            </li>
        </ol>
        </div>
    """, unsafe_allow_html=True)

    st.markdown("""
        <div style='color: #8B0000;'>  <!-- Change this color code to your desired color -->
        <strong>Database Used 🗃️:</strong> project1.sql 
        </div>
    """, unsafe_allow_html=True)

# -------------------------------- PAGE 2: Data Visualization --------------------------------
elif page == "Check post Data Visualization 📊 ":
    st.title("Check Post Data Visualizer 👮🚨")
    
    # Get distinct country names from the database
    country_name = get_data("SELECT DISTINCT country_name FROM check_post")["country_name"].tolist()
    selected_country = st.selectbox("Select Country", country_name)
    
    date_option = st.radio("Filter By:", ["Stop Date", "Entire Month"])
    
    if date_option == "Stop Date":
        selected_date = st.date_input("Choose a Date")
        query = "SELECT * FROM check_post WHERE country_name = %s AND stop_date = %s"
        df = get_data(query, params=(selected_country, selected_date.strftime("%Y-%m-%d")))
    else:
        selected_month = st.selectbox("Select Month", range(1, 13))  # Allow for all months
        query = "SELECT * FROM check_post WHERE country_name = %s AND MONTH(stop_date) = %s"
        df = get_data(query, params=(selected_country, selected_month))

    if not df.empty:
        st.write("### Check Post Data", df)

        # Select feature to visualize
        feature = st.selectbox("Select Feature to Visualize", 
            ["stop_date", "driver_gender", "driver_age", "driver_race", "violation", "stop_outcome", "vehicle_number"])

        # Visualization
        st.write(f"### {feature} Trends")
        plt.figure(figsize=(10, 5))
        sns.lineplot(data=df, x="country_name", y=feature, marker="o", label=feature)
        plt.xticks(rotation=45)
        st.pyplot(plt)

        # Create a bar graph
        plt.figure(figsize=(10, 5))
        sns.histplot(data=df, x="violation", bins=10, kde=True, color='blue')
        plt.title('Violation Count Distribution')
        plt.xlabel('Violations')
        plt.ylabel('Count')
        st.pyplot(plt)
        plt.clf()
        # -------------------------------- PAGE 3: Prediction --------------------------------
elif page == "Prediction":
    st.title("Prediction and outcome")
    df = get_data("SELECT * FROM check_post")

    st.markdown("Fill in the details below to get a predication of the stop outcome based on the existing data.")
    st.header(" Predict outcome and violation")


        # Form for prediction
    with st.form("New_log_form"):  
            stop_date = st.date_input("Stop Date")
            stop_time = st.time_input("Stop Time")
            country_name = st.selectbox("Country", ['Canada', 'India', 'USA'])
            driver_gender = st.selectbox("Driver Gender", ["Male", "Female"])
            driver_age = st.slider("Age", min_value=1, max_value=100, step=1)
            driver_race = st.selectbox("Driver Race",['Asian','Hispanic','White','Black']) 
            search_conducted = st.selectbox("Was a search conducted?", ["TRUE", "FALSE"])
            search_type = st.selectbox("Search Type",['Frisk','Vehichle search'])
            drugs_related_stop = st.selectbox("Was it Drug related?", ["TRUE", "FALSE"])
            stop_duration = st.selectbox("Stop Duration", df['stop_duration'].dropna().unique())
            search_outcome=st.selectbox("Search outcome",['Arrest','Warning','Ticket'])  
            vehicle_number = st.text_input("Vehicle Number", df['vehicle_number'].dropna().iloc[0] if not df['vehicle_number'].empty else "")
            timestamp = pd.Timestamp.now()
            submitted = st.form_submit_button("Predict Stop Outcome and Violation")

    if submitted:
            # Convert string inputs to boolean
            search_conducted_bool = True if search_conducted == "TRUE" else False
            drugs_related_stop_bool = True if drugs_related_stop == "TRUE" else False

            filtered_data = df[
                (df['driver_gender'] == driver_gender) &
                (df['driver_age'] == driver_age) &
                (df['search_conducted'] == search_conducted_bool) &
                (df['stop_duration'] == stop_duration) &
                (df['drugs_related_stop'] == drugs_related_stop_bool)
            ]

            if not filtered_data.empty:
                predicted_outcome = filtered_data['stop_outcome'].mode()[0]
                predicted_violation = filtered_data['violation'].mode()[0]
            else:
                predicted_outcome = "warning"
                predicted_violation = "speeding"

            search_text = "A search was conducted" if search_conducted_bool else "No search was conducted"
            drug_text = "Was drug-related 🍻" if drugs_related_stop_bool else "Was not drug-related"

           
            st.markdown(f"""
                **📋🚓  Prediction Summary**
    
                **Predicted Violation:** {predicted_violation} 🚦


                **Predicted Stop Outcome:** {predicted_outcome} 
            
                    A {driver_age}-year-old {driver_gender} driver in {country_name} was stopped at {stop_time.strftime('%I:%M %p')} on {stop_date} 📅.
                    {search_text} 🔍 and the stop {drug_text}.
                    **Stop Duration:** {stop_duration} ⏳,
                    **Vehicle Number:** {vehicle_number} 🚗
                    """)       

            
                
elif page == "SQL Queries":
    st.title("📋 SQL Query Results")
    st.write("## MEDIUM LEVEL ⌨️")
    queries = {
        "1.Total Number of Police Stops": "SELECT COUNT(*) AS total_stops FROM project1.check_post;",
        "2.Count of Stops by Violation Type": "SELECT violation AS violation_type, COUNT(*) AS stop_count FROM project1.check_post GROUP BY violation ORDER BY stop_count;",
        "3.Number of Arrests vs. Warnings": "SELECT stop_outcome,COUNT(*) AS count FROM project1.check_post WHERE stop_outcome IN ('Arrest', 'Warning') GROUP BY stop_outcome ORDER BY count DESC;",
        "4.Average Age of Drivers Stopped": "SELECT AVG(driver_age) AS average_driver_age FROM  project1.check_post;",
        "5.Top 5 Most Frequent Search Types": "SELECT search_type, COUNT(*) AS frequency FROM project1.check_post GROUP BY search_type ORDER BY frequency DESC LIMIT 5;",
        "6.Count of Stops by Gender": "SELECT driver_gender, COUNT(*) AS stop_count FROM project1.check_post GROUP BY driver_gender ORDER BY stop_count DESC;",
        "7.Most Common Violation for Arrests": "SELECT violation, COUNT(*) AS arrest_count FROM project1.check_post WHERE stop_outcome = 'Arrest' GROUP BY violation ORDER BY arrest_count LIMIT 1;",
        "8.Average Stop Duration for Each Violation": "SELECT violation, AVG(stop_duration) AS average_duration FROM project1.check_post GROUP BY violation;",
        "9.Number of Drug-Related Stops by Year": "SELECT EXTRACT(YEAR FROM stop_date) AS year, COUNT(*) AS drug_stop_count FROM project1.check_post WHERE drugs_related_stop GROUP BY year ORDER BY year;",
        "10.Drivers with the Highest Number of Stops": "SELECT driver_gender, COUNT(*) AS stop_count FROM project1.check_post GROUP BY driver_gender ORDER BY stop_count DESC LIMIT 10;",
        "11.Number of Stops Conducted at Night(Between 10 PM - 5 AM)": "SELECT COUNT(*) AS night_stops FROM project1.check_post WHERE EXTRACT(HOUR FROM stop_time) >= 22 OR EXTRACT(HOUR FROM stop_time) < 5;",
        "12.Number of Searches Conducted by Violation Type": "SELECT violation, COUNT(*) AS search_count FROM project1.check_post WHERE search_conducted = 'true' GROUP BY violation;",
        "13.Arrest Rate by Driver Gender": "SELECT driver_gender,COUNT(CASE WHEN stop_outcome = 'Arrest' THEN 1 END) AS arrest_count,COUNT(*) AS total_stops,(COUNT(CASE WHEN stop_outcome = 'Arrest' THEN 1 END) * 1.0 / COUNT(*)) * 100 AS arrest_rate FROM project1.check_post GROUP BY driver_gender;",
        "14.Violation Trends Over Time (Monthly Count of Violations)": "SELECT DATE_FORMAT(stop_date,'%Y-%m-01') AS month,violation,COUNT(*) AS violation_count FROM project1.check_post GROUP BY month, violation ORDER BY month, violation;",
        "15.Most Common Stop Outcomes for Drug-Related Stops": "SELECT stop_outcome, COUNT(*) AS outcome_count FROM project1.check_post WHERE drugs_related_stop = true GROUP BY stop_outcome ORDER BY outcome_count DESC;"
       }    
    selected_query = st.selectbox("Select a Query to Run", list(queries.keys()))
    query_result = get_data(queries[selected_query])
    
    st.write("### Run Query:")
    st.dataframe(query_result)
    
    st.write("## COMPLEX LEVEL🧠")
    queries = {
        "1.Yearly Breakdown of Stops and Arrests by Country": "SELECT country_name,YEAR(stop_date) AS year,COUNT(*) AS total_stops,COUNT(CASE WHEN stop_outcome = 'Arrest' THEN 1 END) AS total_arrests FROM project1.check_post GROUP BY country_name, year ORDER BY country_name, year;",
        "2.Driver Violation Trends Based on Age and Race": "SELECT driver_age,driver_race,violation,COUNT(*) AS violation_count FROM project1.check_post GROUP BY driver_age, driver_race,violation ORDER BY driver_age, driver_race, violation_count DESC;",
        "3.Time Period Analysis of Stops": "SELECT DATE_FORMAT(stop_date, '%Y-%m') AS month_year,COUNT(*) AS total_stops FROM project1.check_post GROUP BY month_year ORDER BY month_year;",
        "4.Correlation Between Age, Violation, and Stop Duration": "SELECT driver_age, violation, AVG(stop_duration) AS average_duration FROM project1.check_post GROUP BY driver_age, violation ORDER BY driver_age, violation;",
        "5.Violations with High Search and Arrest Rates": "SELECT violation,COUNT(*) AS total_stops,COUNT(CASE WHEN search_conducted = 'true' THEN 1 END) AS total_searches,COUNT(CASE WHEN stop_outcome = 'Arrest' THEN 1 END) AS total_arrests,(COUNT(CASE WHEN stop_outcome = 'Arrest' THEN 1 END) * 1.0 / COUNT(*)) AS arrest_rate,(COUNT(CASE WHEN search_conducted = 'true' THEN 1 END) * 1.0 / COUNT(*)) AS search_rate FROM project1.check_post GROUP BY violation HAVING arrest_rate > 0.1 OR search_rate > 0.1 ORDER BY arrest_rate DESC;",
        "6.Driver Demographics by Country": "SELECT country_name,AVG(driver_age) AS average_age,COUNT(CASE WHEN driver_gender = 'M' THEN 1 END) AS male_count,COUNT(CASE WHEN driver_gender = 'F' THEN 1 END) AS female_count,COUNT(CASE WHEN driver_race IS NOT NULL THEN 1 END) AS total_drivers FROM project1.check_post GROUP BY country_name ORDER BY country_name;",
        "7.Top 5 Violations with Highest Arrest Rates": "SELECT violation,COUNT(*) AS total_stops,COUNT(CASE WHEN stop_outcome = 'Arrest' THEN 1 END) AS total_arrests,(COUNT(CASE WHEN stop_outcome = 'Arrest' THEN 1 END) * 1.0 / COUNT(*)) AS arrest_rate FROM project1.check_post GROUP BY violation ORDER BY arrest_rate DESC LIMIT 5;"    
    }    
    selected_query = st.selectbox("Choose a Query", list(queries.keys()))
    query_result = get_data(queries[selected_query])
    
    st.write("### Query Result:")
    st.dataframe(query_result)

# -------------------------------- PAGE 5: Creator Info --------------------------------
elif page == "Creator Info":
    st.title("👩‍💻 Creator of this Project")
    st.write("""
    **Developed by:** AGILA.R 

    **Skills:** 
         
          > Python
         > SQL
         > Data Analysis 📈
         > Streamlit
         > Pandas🐼  
    """)
```
